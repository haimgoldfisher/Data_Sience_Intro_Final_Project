## a.

It said that $1/125$ of the births are of non-identical twins and $1/300$ are of identical twins. Moreover, the probabilty to give birth to a boy is $1/2$ .  

We know that Elvis had a twin brother, so we care about the probability of being a identical twin, when it given that the baby is a twin and a boy. Such that:

P(identical twins & twin brother) = P(identical twins) $\bullet$ P(both boys|identical twins) = $\frac{1}{2} \bullet \frac{1}{300}$

P(fraternal twins & twin brother) = P(fraternal twins) $\bullet$ P(both boys|fraternal twins) = $\frac{1}{4} \bullet \frac{1}{125}$

That means we can use the conditional probability's formula:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
        
Let A be the identical twin case and let B be the twin and boy case - P(identical twins|twin brother):

$$P(A|B) = \frac{1/2 \bullet 1/300}{(1/2 \bullet 1/300) + (1/4 \bullet 1/125)} = 0.454545$$

##### The probability that Elvis's twin was identical twin is 0.454545 .

# Question 2 :

M&M candies packet from 1994's distribution: 30% Brown, 20% Yellow, 20% Red, 10% Green, 10% Orange, 10% Tan.

M&M candies packet from 1996's distribution: 24% Blue , 20% Green, 16% Orange, 14% Yellow, 13% Red, 13% Brown.

A friend gave me two candies, one from each packet. One candy is yellow and one candy is green.

We need to find the probability that the yellow candy came from the 1994's packet.

We can use the conditional probability's formula:


$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

Let A be the yellow candies from 1994's packet, let B be the yellow candies from 1996's packet.

We know that 20% of the candies from 1994's packet are yellow, and 14% of the candies from 1996's packet are yellow. So that:


$$P(A|B) = \frac{20/100}{(20/100)+(14/100)} = 0.5888235$$

##### The probability that the yellow candy came from the 1994's packet is 0.5888235 .

## a.

We know statistically that one in 10,000 people in the population is sick with swine influenza.
It is known that a medical examination of a swine influenza has a 99% of success rate. It means that the probability that a "sick" person isn't sick at all is 1%.

We need to find the probability that a random person that did the medical examination and found out that he is ill, is truly ill. 

We can use the formula of Bayes' theorem:

$$P(B|A) = \frac{P(A|B) \bullet P(B)}{P(A)}$$


$$P(positive) = ( P(positive|A) \bullet P(A) ) + ( P(positive|B) \bullet P(B) )$$

We know that the medical examination can't be wrong when the person is sick.

$$P(sick|positive) = 1 $$
$$P(sick) = \frac{1}{10,000}$$
$$P(healthy|positive) = \frac {1}{100} $$
$$P(healthy) = \frac{9999}{10,000}$$


$$P(positive) = 1 \bullet \frac {1}{10,000} + \frac {1}{100} \bullet \frac {9999}{10,000} = 0.010099$$

##### The probability that a random person that did the medical examination and found out that he is ill, is truly ill is 0.010099 .
